### 1. 生成数据集

```python
import numpy as np
import torch
from torch.utils import data
from d2l import torch as d2l
```

```python
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = d2l.synthetic_data(true_w, true_b, 1000)
```

### 2. 读取数据集

```python
def load_array(data_arrays, batch_size, is_train=True):
    """构造一个Python数据迭代器"""
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset, batch_size, shuffle=is_train)

batch_size = 10
data_iter = load_array((features, labels), batch_size)
```

```python
# 与 3.2节不同，这里我们使用iter构造Python迭代器，并使用next从迭代器中获取第一项。
next(iter(data_iter))
```

### 3. 定义模型

```python
# nn是神经网络的缩写
from torch import nn

net = nn.Sequential(nn.Linear(2, 1))
```

### 4. 初始化模型参数


```python
net[0].weight.data
net[0].bias.data

net[0].weight.data.normal_(0, 0.01)
net[0].bias.data.fill_(0)
```

### 5. 定义损失函数

```python
#默认情况下，他返回所有样本损失的平均值
loss = nn.MSELoss()
```

### 6. 定义优化算法

```python
trainer = torch.optim.Adam(net.parameters(), lr = 0.03)
```

### 7. 训练

```python
num_epochs = 3
for epoch in range(num_epochs):
    for X, y in data_iter:
        l = loss(net(X) , y)
        trainer.zero_grad()
        l.backward()
        #print(net[0].weight.grad)
        trainer.step()
    l = loss(net(features), labels)
    print(f'epoch {epoch + 1}, loss {l:f}')
```

```python
w = net[0].weight.data
print('w的误差估计：', true_w - w.reshape(true_w.shape))
b = net[0].bias.data
print('b的误差估计：', true_b - b)
```

### 练习


```python
#生成数据集
true_w = torch.tensor([-3, 7.2])
true_b = 3.6
features, labels = d2l.synthetic_data(true_w, true_b, 1500)
#读取数据集
def load_array(data_arrays, batch_size, is_train=True):
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset, batch_size, shuffle=is_train)
batch_size = 15
data_iter = load_array((features, labels), batch_size)
#定义模型、损失、优化器
from torch import nn
net = nn.Sequential(nn.Linear(2, 1))
loss = nn.MSELoss()
trainer = torch.optim.SGD(net.parameters(), lr=0.03)
#初始化模型参数
net[0].weight.data.normal_(0, 0.01)
net[0].bias.data.fill_(0)
#训练 3 个 epoch
num_epochs = 3
for epoch in range(num_epochs):
    for X, y in data_iter:
        l = loss(net(X), y)
        trainer.zero_grad()
        l.backward()
        trainer.step()
    l = loss(net(features), labels)
    print(f'epoch {epoch+1}, loss {l:f}')
#打印学到的参数
w = net[0].weight.data
print('w的估计误差：', true_w - w.reshape(true_w.shape))
b = net[0].bias.data
print('b的估计误差：', true_b - b)
```
