## 索引和切片

```python
x = torch.arange(10)
print(x)
print()
print(x[0])
print(x[-1])
print(x[2:5])
print(x[:3])
print(x[3:])
print(x[::2])
```

```python
x = torch.arange(12).reshape(3, 4)
print(x)
print(x[1, 2])
#核心语法：X[行, 列]，用逗号分隔不同维度的索引。

Y = torch.arange(24).reshape(2,3,4)
print(Y)
print(Y[:, 0, :])
```

```python
#修改张量的值
X = torch.arange(12).reshape(3,4)
X[1, 2] = 99
print(X)
X[0, :]=0
print(X)
X[:, 0] = torch.tensor([100,200,300])
print(X)
```

## 张量运算
形状相同的张量

```python

x = torch.tensor([1.0,2,4,8])
y = torch.tensor([2,2,2,2])

print(x + y)
print(x - y)
print(x / y)
print(x * y)
print(x ** y)
torch.exp(x)
```

```python
#张量拼接
X = torch.arange(12, dtype=torch.float32).reshape((3,4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print(torch.cat((X, Y),dim=0))
torch.cat((X, Y), dim=1)
```

```python
X > Y
```

## 节省内存

```python
before = id(Y)
print(before)
Y = Y + X
print(id(Y))
```

```python
#用切片表示法将操作的结果分配给原先分配的数组：执行原地操作
Z = torch.zeros_like(Y)
print("id(Z):", id(Z))
Z[:] = X+Y
print("id(Z):", id(Z))
```

```python
before = id(X)
X += Y
id(X) == before
```

```python
#这三种等价：
X[:] = X + Y
X += Y
X.add_(Y)
```
