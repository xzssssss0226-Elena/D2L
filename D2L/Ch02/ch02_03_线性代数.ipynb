```python
#非降维求和
sum_A = A.sum(axis=1,keepdims=True)
print(sum_A)
```

```python
#由于sum_A在对每行进行求和后仍保持两个轴，我们可以通过广播将A除以sum_A。

A/sum_A
```

```python
#沿某个轴计算A元素的累积总和， 比如axis=0（按行计算），
#可以调用cumsum函数。 此函数不会沿任何轴降低输入张量的维度。
print(A)
A.cumsum(axis=0)
```

```python
C = torch.arange(24).reshape(2,3,4)
print(C)
C.cumsum(dim=1)
```

## 检测题

```python
A = torch.randn(3,4)
print(torch.sum((A.T).T == A) / torch.numel(A) == 1)
print(((A.T).T == A).sum())
print(torch.sum((A.T).T == A))

# A + A.T ==(A +A.T).T
```

```python
A = torch.randn(4,4)
M = A + A.T
print((M == M.T).all())
#.all()更简洁，直接判断“是否所有元素都是true”
```
