```python
import os

os.makedirs(os.path.join('..', 'data'), exist_ok=True)
data_file = os.path.join('..', 'data', 'house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('NumRooms,Alley,Price\n')  # 列名
    f.write('NA,Pave,127500\n')  # 每行表示一个数据样本
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
```

## 读取数据集

```python
# 如果没有安装pandas，只需取消对以下行的注释来安装pandas
!pip install pandas
import pandas as pd

data = pd.read_csv(data_file)
print(data)
```

## 处理缺失值

```python
#插值法
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2] #iloc=按位置取
inputs = inputs.fillna(inputs.mean(numeric_only=True)) #默认算列
print(inputs)
```

```python
#处理字符串（独热编码）
inputs = pd.get_dummies(inputs, dummy_na=True)
print(inputs)
```

```python
#转成张量
import torch
X = torch.tensor(inputs.to_numpy(dtype=float))
y = torch.tensor(outputs.to_numpy(dtype=float))
X, y
```

```python
N_na = data.isna().sum()
print(N_na)
print(N_na.idxmax())
#Series = 带标签的一维数据。
#可以理解成"一列数据 + 每个数据有个名字"。Pandas DataFrame多个Series拼起来。
```

```python
data.drop(columns="Alley") #不改变原data，而是返回一个新的DataFrame
```

```python
data_new = data.drop(columns=data.isna().sum().idxmax())
print(data_new)
```
